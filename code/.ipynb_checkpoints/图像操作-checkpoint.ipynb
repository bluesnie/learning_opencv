```python
def cv_show(name, img):
    for item in range(len(img)):
        cv2.imshow(name + str(item), img[item])
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## 数据读取-视频
- cv2.VideoCapture可以捕获摄像头，用数字来控制不同的设备，例如0,1。
- 如果是视频文件，直接指定好路径即可。

```python
vc = cv2.VideoCapture("../img/test.mp4")
```

```python
# 检查是否打开正确
if vc.isOpened():
    open, frame = vc.read()  # vc.read() 读取一帧，如果读取到了open=True,frame就是当前帧的图像
else:
    open = False
```

```python
while open:
    ret, frame = vc.read()
    if frame is None:
        break
    if ret is True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # 3维通道转灰度
        cv2.imshow('ret', gray)
        if cv2.waitKey(5) & 0xFF == 27:   # 5s播放，27是一个退出键（esc）
            break
vc.release()
cv2.destroyAllWindows()
```

## 边界填充
- BORDER_REPLICATE：复制法，也就是复制最边缘像素。
- BORDER_REFLECT：反射法，对感兴趣的图像中的像素在两边进行复制例如：fedcba|abcdefgh|hgfedcb
- BORDER_REFLECT_101：反射法，也就是以最边缘像素为轴，对称，gfedcb|abcdefgh|gfedcba
- BORDER_WRAP：外包装法cdefgh|abcdefgh|abcdefg
- BORDER_CONSTANT：常量法，常数值填充。

```python
top_size, bottom_size,left_size,right_size = (50, 50, 50, 50)
replicate = cv2.copyMakeBorder(img, top_size,bottom_size,left_size,right_size, cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img, top_size,bottom_size,left_size,right_size, cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img, top_size,bottom_size,left_size,right_size, cv2.BORDER_REFLECT101)
wrap = cv2.copyMakeBorder(img, top_size,bottom_size,left_size,right_size, cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img, top_size,bottom_size,left_size,right_size, cv2.BORDER_CONSTANT, value=0)
```

```python
plt.subplot(231), plt.imshow(img, 'gray'), plt.title('ORIGINAL')
plt.subplot(232), plt.imshow(replicate, 'gray'), plt.title('REPLICATE')
plt.subplot(233), plt.imshow(reflect, 'gray'), plt.title('REFLECT')
plt.subplot(234), plt.imshow(reflect101, 'gray'), plt.title('REFLECT_101')
plt.subplot(235), plt.imshow(wrap, 'gray'), plt.title('WRAP')
plt.subplot(236), plt.imshow(constant, 'gray'), plt.title('CONSTANT')

plt.show()
```

## 数值计算
- 因为颜色数值为：0-255，所以计算的时候超过后会变成： 计算后的数值 % 256（取模）
- 因此计算需要转换成float后，之后小于0的转换为0，大于255的转换为255

```python
cal_img = img.copy()

cal_img1 = cal_img + 100

cal_img2 = cal_img.astype(np.float) + 100
```

```python
cal_img
```

```python
cal_img1
```

```python
cal_img2
```

```python
cal_img2.clip(0,255).astype(np.uint)
```

## 图像融合

```python
img_cat = cv2.imread("../img/cat.jpg")
img_dog = cv2.imread("../img/dog.jpg")
```

```python
img_cat.shape
```

```python
img_dog.shape
```

```python
img_dog_resize = cv2.resize(img_dog, (500, 414))
img_dog_resize.shape
```

```python
cv_show('ret', [img_cat, img_dog, img_dog_resize])
```

```python
# R = αx + βy + b；其中α和β是对应的权重，哪个值大就更明显，b是亮度级上提亮
res = cv2.addWeighted(img_cat, 0.4, img_dog_resize, 0.6, 0)
plt.imshow(res)
```

```python
# 图像大小改变，按不同比例放缩
res = cv2.resize(img, (0, 0), fx=5, fy=5)
print("大小", res.shape)
plt.imshow(res)
```
